```python
#Algorithms & design techniques
#Key idea: short, canonical examples: binary search, merge sort, greedy, DP (knapsack), Dijkstra
# binary search
def bsearch(a,x):
    lo,hi=0,len(a)-1
    while lo<=hi:
        m=(lo+hi)//2
        if a[m]==x: return m
        if a[m]<x: lo=m+1
        else: hi=m-1
    return -1

# merge sort (divide-and-conquer)
def merge_sort(a):
    if len(a)<=1: return a
    m=len(a)//2
    L=merge_sort(a[:m]); R=merge_sort(a[m:])
    i=j=0; res=[]
    while i<len(L) and j<len(R):
        if L[i]<R[j]: res.append(L[i]); i+=1
        else: res.append(R[j]); j+=1
    return res+L[i:]+R[j:]

# greedy: activity selection (end times)
def activity_sel(intervals):
    intervals.sort(key=lambda x:x[1])
    last=-1; res=[]
    for s,e in intervals:
        if s>=last: res.append((s,e)); last=e
    return res

# DP: 0/1 knapsack (weights, values small example)
def knap(w, W, val):
    n=len(w)
    dp=[[0]*(W+1) for _ in range(n+1)]
    for i in range(1,n+1):
        for cap in range(W+1):
            dp[i][cap]=dp[i-1][cap]
            if w[i-1]<=cap:
                dp[i][cap]=max(dp[i][cap], val[i-1]+dp[i-1][cap-w[i-1]])
    return dp[n][W]
```

```python
#Graph utilities with DFS and BFS (runnable Python)
#Below is a compact, self-contained Python script that implements an adjacency-list graph, recursive DFS, iterative DFS, and BFS, with a small demo graph and printed traversal order. Save as graph_traversals.py and run with python3 graph_traversals.py.
#!/usr/bin/env python3
# graph_traversals.py
# Compact graph utilities: adjacency list, BFS, recursive DFS, iterative DFS.

from collections import deque
from typing import Dict, List, Set

Graph = Dict[int, List[int]]

def add_edge(g: Graph, u: int, v: int, undirected: bool = True) -> None:
    """Add edge u->v to graph; if undirected, also add v->u."""
    g.setdefault(u, []).append(v)
    if undirected:
        g.setdefault(v, []).append(u)

def bfs(g: Graph, start: int) -> List[int]:
    """Breadth-first search returning visitation order."""
    vis: Set[int] = set()
    order: List[int] = []
    q = deque([start])
    vis.add(start)
    while q:
        u = q.popleft()
        order.append(u)
        for w in g.get(u, []):
            if w not in vis:
                vis.add(w)
                q.append(w)
    return order

def dfs_recursive(g: Graph, start: int) -> List[int]:
    """Recursive DFS returning visitation order."""
    vis: Set[int] = set()
    order: List[int] = []
    def dfs(u: int):
        vis.add(u)
        order.append(u)
        for w in g.get(u, []):
            if w not in vis:
                dfs(w)
    dfs(start)
    return order

def dfs_iterative(g: Graph, start: int) -> List[int]:
    """Iterative DFS using an explicit stack (preorder)."""
    vis: Set[int] = set([start])
    order: List[int] = []
    stack: List[int] = [start]
    while stack:
        u = stack.pop()
        order.append(u)
        # push neighbors in reverse to visit in same order as recursive DFS
        for w in reversed(g.get(u, [])):
            if w not in vis:
                vis.add(w)
                stack.append(w)
    return order

def demo():
    # Build a sample undirected graph:
    # 0 - 1 - 2
    # |   |
    # 3 - 4
    g: Graph = {}
    add_edge(g, 0, 1)
    add_edge(g, 1, 2)
    add_edge(g, 0, 3)
    add_edge(g, 1, 4)
    add_edge(g, 3, 4)

    print("Adjacency list:")
    for u in sorted(g):
        print(f" {u}: {g[u]}")

    start = 0
    print("\nBFS order:", bfs(g, start))
    print("DFS recursive order:", dfs_recursive(g, start))
    print("DFS iterative order:", dfs_iterative(g, start))

if __name__ == "__main__":
    demo()
```
